### functions as arguments

```python
def f(x): return x**2
```

```python
def g(x): return np.cos(x**3)
```

```python
def avg(x,f):
    # computes the average of the values of f at the nodes in x
    # note f is a function
    
    m=np.mean( f(x) )
    
    return m
```

```python
x=np.arange(10)
avg(x,f)
```

```python
avg(x,g)
```

```python
avg(x,np.sin)
```
