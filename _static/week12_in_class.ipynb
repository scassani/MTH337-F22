```python
si=string.printable.find(" ") # space index
ei=string.printable.find("!") # ! index
string.printable
```

```python
for char in verne:
    if char not in string.printable:
        verne=verne.replace(char," ")
```

```python
vsplit=verne.lower().split()
print(vsplit[:20])
```

```python
wc={} # word count dictionary
```

```python
for word in vsplit:
    if word in wc:
        wc[word]+=1
    else:
        wc[word]=1

wc
```

```python
list(wc.items())[:10]
```

```python
wcl=list(wc.items())
sorted(wcl)
```

### Regular expressions

```python
import re
```

```python
url="https://www.gutenberg.org/files/103/103-0.txt"

verne = requests.get(url).text
verne
```

```python
re.findall(r'\b\w\w\w\w\w\b',verne)[:20] # \w any word char (numbers, letters, underscore)
# \b word delimiter
```

```python
re.findall(r'\b\w{5}\b',verne)[:20]
```

```python
re.findall(r'\w{5}',verne)[:20]
```

```python
re.findall(r'\b\d{4}\b',verne)[:20] # \d is for digits
```

```python
re.findall(r'\b\d{3}2\b',verne)[:20]
```

find all words of 5 chararcters, starting with `i` ending with `t`

```python
re.findall(r'\bi\w{3}t\b',verne)[:20]
```

find all words starting with `i` ending with `t`

```python
print(re.findall(r'\bi\w+t\b',verne)) # `+` matches 1 or more times
```

```python
print(re.findall(r'\bi\w*t\b',verne)) # `*` mathces 0 or more times
```

```python
print(re.findall(r'\bi\w{3,6}t\b',verne)) # between 3 and 6 times
```

```python
print(re.findall(r'\bi\w{3,}t\b',verne)) # more or = than 3 times
```

```python
print(re.findall(r'\bi\w{,5}t\b',verne)) # less or = than 5 times
```

Find all words starting with `a` or `b` or `c` and ending with `t`

`[]` indicates a list of chars that can be matched

```python
a=re.findall(r"\b[abc]\w*t\b",book)
print(a)
```

```python
a=re.findall(r"\b[A-Z]\w*t\b",book)
print(a)
```

```python
a=re.findall(r"\b[A-Zb-g]\w*t\b",book)
print(a)
```

```python
a=re.findall(r"\b[^abc]\w*t\b",book) # starting with everything but a,b,c
print(a)
```

```python
a=re.findall(r"\b[^abc\W]\w*t\b",book) # starting with everything but a,b,c,
                # and must be a character (\W is the opposite of \w)
print(a)
```

```python
a=re.findall(r"\W\b",book) # starting with everything but a,b,c
print(a)
```

```python
a=re.findall(r"\b[ABCabc]\w{11,12}\b",book) 
print(a)
```

```python
i=book.find("BROBDINGNAG_")
```

```python
book[i-20:i+20]
```

Find all lines with a number in it

```python
a=re.findall(r"\b\d+\.{,1}\d*\b",book) 
print(a)
```

```python
a=re.findall(r"\b\d+[\.:]{,1}\d*\b",'12.3 is nicer than 0.123 and 123 at 12:30') 
print(a)
```
