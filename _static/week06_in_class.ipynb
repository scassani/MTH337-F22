### subplots 

`plt.subplot(r,c,p)`

```python
x=np.linspace(0,2*np.pi,201)
y=np.sin(x)
z=np.cos(x)
w=np.tan(x)
v=1/np.tan(x)
```

```python
plt.figure(figsize=(10,5))
plt.subplots_adjust(wspace=0.5,hspace=0.2)
plt.subplot(1,3,1)
plt.plot(x,y)
plt.title('sinx')
plt.subplot(1,3,2)
plt.plot(x,z)
plt.subplot(1,3,3)
plt.plot(x,w)
plt.ylim([-5,5])
plt.suptitle('Trig functions');
```

```python
plt.figure(figsize=(10,10))
plt.subplots_adjust(wspace=0.5,hspace=0.2)
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('sinx')
plt.subplot(2,2,2)
plt.plot(x,z)
plt.subplot(2,2,3)
plt.plot(x,w)
plt.ylim([-5,5])
plt.subplot(2,2,4)
plt.plot(x,v)
plt.ylim([-5,5])
plt.suptitle('Trig functions');
```
