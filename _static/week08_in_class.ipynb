```python
img=plt.imread('img.jpeg')
```

```python
# original
plt.subplot(2,2,1)
plt.imshow(img)
# red scale
r=img.copy() 
r[:,:,1:]=0 # zero out green and blue
plt.subplot(2,2,2)
plt.imshow(r)
# green scale
g=img.copy() 
#g[:,:,0]=0 # zero out red and blue
#g[:,:,2]=0
g[:,:,[0,2]]=0 
plt.subplot(2,2,3)
plt.imshow(g)
# blue scale
b=img.copy() 
b[:,:,:2]=0 
plt.subplot(2,2,4)
plt.imshow(b);
```

```python
# dictionary
palette={'K' : [16, 16, 16],'DB' : [32, 32, 96],'DG' : [0, 56, 32],'HG' : [40, 88, 0],'Y' : [232, 192, 0]}
```

**Ex.** given a 20x20 image. Use the first 3 colors in you palette to build vertical stripes of width 11,4,5 respectively

```python
i=np.zeros((20,20,3),dtype=int)
i[:,0:11,:]=palette["K"]
i[:,11:15,:]=palette["DB"]
i[:,15:20,:]=palette["DG"]
plt.imshow(i);
```

```python
plt.imshow(np.transpose(i,[1,0,2]))
```
